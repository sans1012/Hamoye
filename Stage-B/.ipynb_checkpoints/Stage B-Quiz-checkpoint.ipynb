## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

## Load the csv

```python
df= pd.read_csv("energydata_complete.csv") 
```

### Removing unrequired columns

```python
df.drop(columns=['date', 'lights'], inplace=True)
```

```python
df.columns
```

## Data Preprocessing

```python
scalar= MinMaxScaler()
```

```python
normalised_df= pd.DataFrame(data= scalar.fit_transform(df), columns=df.columns)
```

```python
X= normalised_df.drop(['Appliances'], axis=1)
y=normalised_df['Appliances']
```

## Train test

```python
X_train,X_test,y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=42)
```

## Linear Regression

```python
##Linear Regression
lr=LinearRegression()
lr.fit(X_train, y_train)
y_pred=lr.predict(X_test)


accuracy_matrix=pd.DataFrame(data= [round(mean_absolute_error(y_test, y_pred),2),
                        round(mean_squared_error(y_test, y_pred),2),
                       round(np.sum(np.square(y_test- y_pred)),2), 
                       round(r2_score(y_test, y_pred),2),
                       round(np.sqrt(mean_squared_error(y_test, y_pred)),3)], 
                 index= ["Mean Absolute Error",
                        "Mean Square Error",
                         "Residual Error(RSS)",
                         "R2 Score (CoD)", "Root mean Square"],
                 columns= ['Linear Regression'])
```

## Ridge Regression

```python
## Ridge Regression
ridge=Ridge()
ridge.fit(X_train, y_train)


accuracy_matrix['Ridge Regression']=[round(mean_absolute_error(y_test , ridge.predict(X_test)), 2),
                                     round(mean_squared_error(y_test , ridge.predict(X_test)), 2),
                                     round(np.sum(np.square(y_test - ridge.predict(X_test))), 2),
                                     round(r2_score(y_test, ridge.predict(X_test)), 2), 
                                     round(np.sqrt(mean_squared_error(y_test, ridge.predict(X_test))), 3)]
```

## Lasso Regression

```python
## Lasso
lasso=Lasso(alpha=0.001)
lasso.fit(X_train, y_train)



accuracy_matrix['Lasso']=[round(mean_absolute_error(y_test , lasso.predict(X_test)), 2),
                          round(mean_squared_error(y_test , lasso.predict(X_test)), 2),
                          round(np.sum(np.square(y_test - lasso.predict(X_test))), 2),
                          round(r2_score(y_test, lasso.predict(X_test)), 2),
                          round(np.sqrt(mean_squared_error(y_test,lasso.predict(X_test))), 3)]
```

```python
# Training set of the T2 and T6 columns 
train_x = X_train[['T2']]
train_y = X_train[['T6']]

# Testing set of the T2 and T6 column 
test_x = X_test[['T2']]
test_y = X_test[['T6']]

#LinearRegression model
model = LinearRegression()

# Fit the model with training set
model.fit(train_x, train_y)

p = model.predict(test_x)


#Compute the r2_score on test set
R2_score = round(r2_score(test_y, p), 3)
R2_score
```

```python
##weights

def get_weights(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_model_weights= get_weights(lr, X_train, "Linear_Model_Weight" )
ridge_weights_df= get_weights(ridge, X_train, "Ridge Weight")
lasso_weights_df= get_weights(lasso, X_train, "Lasso Weight")
```

```python
final_weights= pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights= pd.merge(final_weights, lasso_weights_df, on= "Features")
```

```python
final_weights.sort_values("Linear_Model_Weight")
```
