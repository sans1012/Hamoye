## Data Visualization

```python
df['Total']=df['Y2014']+df['Y2015']+df['Y2018']+df['Y2016']+df['Y2017']
Population=df[df['Element']=='Total Population - Both sexes']
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=Population['Area'],y=Population['Y2014'],palette='Set2' )
plt.xticks(rotation=90);
plt.title("Total Polulation")
plt.show()
```

Maximum population is at nigeria while lowest is in Sao Tome and Principe

```python
sns.lineplot(df['Y2017'],df['Y2018'])
plt.title("relation in sale in 2017 and 2018")
```

there is a linear relationship in the prouction over the years

```python
sns.histplot(df, x='Y2015', bins=10, kde=True, color='Grey', fill=True)
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x=df['Element'],palette='bright')
plt.xticks(rotation=90);
plt.title("Count of different elements")
plt.show()
```

Highest: Domestic Supply

Lowest: Tourist Consumption

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Area', data=df,palette='Set2')
plt.xticks(rotation=90);
plt.title("Number of different countries")
plt.show()
```

Area with max supply and other facilities is Kenya then Egypt and lowest is Euthopia

```python
a=df.groupby('Item').sum()
a=a.sort_values(by='Total', ascending=False)[:10]


colors = sns.color_palette('Paired')[0:10]
l=['Population','Cereals - Excluding Beer','Starchy Roots','Cassava and products',
                          'Sugar Crops','Maize and products', 'Fruits - Excluding Wine', 'Sugar cane', 
                          'Milk - Excluding Butter' ,'Vegetables']
plt.figure(figsize=(7,9))

plt.pie(a['Total'],labels=l ,colors = colors, autopct='%.0f%%')
plt.title("Top ten Items ")
plt.show()
```

Highest Element is Population 

```python
df_item=df.groupby('Element').sum().sort_values(by='Total', ascending='False')
df_item=df_item[-10:]

plt.figure(figsize=(10,6))
sns.barplot(data=df_item, x=['Losses', 'Export Quantity', 'Feed', 'Import Quantity', 'Processing', 
                              'Food supply (kcal/capita/day)', 'Total Population - Both sexes' ,'Food',
                                'Production', 'Domestic supply quantity '], y='Total')
plt.xticks(rotation=90)
plt.title("Demonstration of various elements")
plt.show()
```

Highest element is Demestic Supply Quantity (n overall all the years).
